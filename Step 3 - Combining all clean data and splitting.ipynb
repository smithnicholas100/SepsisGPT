# Run all (except exporting files)

```python
###   LOAD DATA   ###
pos = pd.read_csv(csv_file_path+'/clean_sepsis_pos.csv')
neg  = pd.read_csv(csv_file_path+'/clean_sepsis_neg.csv')
# get rid of Unnamed: 0 column
pos = pos.drop('Unnamed: 0', axis=1)
neg = neg.drop('Unnamed: 0', axis=1)
```

```python
pos.head()
```

```python
# Add a new column named 'New_Column' with 'Yes' as the value in every cell
pos['target'] = 'Yes'

# Display the DataFrame with the new column
pos.head()
```

```python
# Add a new column named 'New_Column' with 'Yes' as the value in every cell
neg['target'] = 'No'

# Display the DataFrame with the new column
neg.head()
```

```python
# Take a random sample with the same number of rows as pos
neg_sampled = neg.sample(n=len(pos), replace=False, random_state=42)

# Display the sampled DataFrame
neg_sampled.info()
```

```python
balanced = pd.concat([neg_sampled,pos], ignore_index=True)
balanced.info()
```

```python
balanced.head(11)
```

```python
# Shuffle the rows of the DataFrame
shuffled = balanced.sample(frac=1, random_state=42)

shuffled.head(20)
```

```python
# Calculate the number of rows for each split
total_rows = len(shuffled)
train_size = int(0.70 * total_rows)  # 70% for training
validate_size = int(0.15 * total_rows)  # 15% for validation
test_size = total_rows - train_size - validate_size  # Remaining for testing

# Split the DataFrame into train, validate, and test DataFrames
train = shuffled[:train_size]
validate = shuffled[train_size:train_size + validate_size]
test = shuffled[train_size + validate_size:]

# Display the sizes of train, validate, and test sets
train.info()
validate.info()
test.info()
```

```python
test['target'].value_counts()
```

```python
assert len(shuffled) == len(train) + len(validate) + len(test), "ssomething went wrong"
```

```python
# restart the index

train = train.reset_index(drop=True)
validate = validate.reset_index(drop=True)
test = test.reset_index(drop=True)
```
